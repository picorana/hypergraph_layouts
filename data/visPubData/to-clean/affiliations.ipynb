```python
import pandas as pd
import numpy as np
```

```python
fp = "authors-affiliations-2021-toclean.csv"
```

```python
df = pd.read_csv(fp)
```

```python
df.head()
```

```python
len(df)
```

```python
cleaned_affiliations = df.CleanedAffiliation
```

```python
np.count_nonzero(pd.isna(cleaned_affiliations))
```

```python
cleaned_set = set(cleaned_affiliations)
```

```python
len(cleaned_set)
```

```python
cleaned_set
```

```python
with open('cleaned_aff', "w") as fp:
    for aff in cleaned_set:
        fp.write(str(aff) + "\n")
```
